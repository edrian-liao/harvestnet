This notebook combines the relabelled positive subdivided csv files and the negative subdivided csv files.

```python
import pandas as pd
import csv
```

```python
df_pos_train = pd.read_csv("../Dataset/train_mod_pos_subdivided.csv")
df_pos_test = pd.read_csv("../Dataset/test_mod_pos_subdivided.csv")
df_neg_train = pd.read_csv("../Dataset/train_buff_neg_subdivided.csv")
df_neg_test = pd.read_csv("../Dataset/test_buff_neg_subdivided.csv")
```

```python
df_merged_train = pd.concat([df_pos_train, df_neg_train])
df_merged_test = pd.concat([df_pos_test, df_neg_test])
```
